#### Marcov chain

```python
import numpy as np
from hmmlearn import hmm
import matplotlib.pyplot as plt
```

```python
# Convert date to datetime object
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'], format='%d/%m/%Y')
```

```python
df_cleaned.info()
```

```python
df_cleaned.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Selecting features
features = df_cleaned[['new_cases_per_100_000', 'new_cases_7_day_rolling_avg', 'new_deaths_7_day_rolling_avg']]
target = df_cleaned['new_deaths_per_100_000']

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Check the scaled features
print(features_scaled[:5])
```

#### Build Model

```python

from hmmlearn import hmm
import numpy as np

# Assume features_scaled and target are already defined

# Number of components (states) you expect in the data
n_components = 4  # Example: 4 states, adjust based on domain knowledge or experimentation

# Initialize Gaussian HMM
model = hmm.GaussianHMM(n_components=n_components, covariance_type="diag", n_iter=100)

# Fit model on the features
model.fit(features_scaled)

# Predict states
states = model.predict(features_scaled)







```

```python
# Add the states to the DataFrame
df_cleaned['state'] = states

# Analyze the average or distribution of `new_deaths_per_100_000` per state
state_analysis = df_cleaned.groupby('state')['new_deaths_per_100_000'].describe()
print(state_analysis)
```
