#### Marcov chain

```python
import numpy as np
from hmmlearn import hmm
import matplotlib.pyplot as plt
```

```python
# Convert date to datetime object
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'], format='%d/%m/%Y')
```

```python
df_cleaned.info()
```

```python
df_cleaned.head()
```

#### Build Model

```python
# Ensure sorting
df_cleaned.sort_values(by=['state','date'], inplace=True)

df_cleaned.head()
```

```python
alabama_df = df_cleaned[df_cleaned['state'] == 'Alabama']
alabama = alabama_df[['state','new_cases_7_day_rolling_avg', 'new_deaths_7_day_rolling_avg', 'cumulative_cases_per_100_000', 'cumulative_deaths_per_100_000']]
alabama.head()
```

```python
states = df_cleaned['state'].unique()

# Dictionary to hold models
state_models = {}
state_predictions = {}

# Train a model for each state
for state in states:
    state_data = df_cleaned[df_cleaned['state'] == state]
    X = state_data[['new_cases_7_day_rolling_avg', 'new_deaths_7_day_rolling_avg', 
                    'cumulative_cases_per_100_000', 'cumulative_deaths_per_100_000']].values

    # Train HMM
    #Low Severity: Few cases and deaths, limited community spread.Moderate Severity: Increasing cases, growing community transmission, and rising deaths.High Severity: High number of cases and deaths, widespread transmission.

    model = hmm.GaussianHMM(n_components=3, covariance_type="diag", n_iter=1000) 

    model.fit(X)
    state_models[state] = model
    
    # Predict states
    predicted_states = model.predict(X)
    state_predictions[state] = predicted_states
```
